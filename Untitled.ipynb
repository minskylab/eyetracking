```
import requests
import json
import io

import numpy as np
from PIL import Image
from matplotlib import pyplot as plt
%matplotlib inline
```

```
info = requests.get("https://eyetracking.minsky.cc/api/v1/info")
images = json.loads(info.content)["images"]
```

```
BASE_IMAGES_URL = "https://minskypublic.nyc3.digitaloceanspaces.com"
sizes = []
for img in images:
    url = BASE_IMAGES_URL + "/" + img
    r = requests.get(url)
    image = Image.open(io.BytesIO(r.content))
    sizes.append(image.size)
sizes = np.array(sizes)
```

```
widths = sizes[:, 0]
heights = sizes[:, 1]
plt.hist(widths, 10)
```

```
plt.hist(heights, 10)
```

```
counts, bins = np.histogram(widths)
indexes = counts.argsort()
wMean1 = bins[indexes[-1]]
wMean2 = bins[indexes[-2]]

counts, bins = np.histogram(heights)
indexes = counts.argsort()
hMean1 = bins[indexes[-1]]
hMean2 = bins[indexes[-2]]
```
